```python
clean_earthquake_copy = clean_earthquake.copy()

# Drop columns by name
clean_earthquake_copy = clean_earthquake_copy.drop(["time_utc", "place", "url"], axis=1, errors='ignore')

# Encode 'id'
id_mapping = {v: i for i, v in enumerate(clean_earthquake_copy['id'].unique(), 1)}
clean_earthquake_copy['id_encoded'] = clean_earthquake_copy['id'].map(id_mapping)

# Encode 'magtype'
magtype_mapping = {
    'mb': 1,
    'ml': 2,
    'mw': 3,
    'mwr': 4,
    'mwb': 5,
    'mwc': 6,
    'uk': 0
}
clean_earthquake_copy['magtype_encoded'] = clean_earthquake_copy['magtype'].map(magtype_mapping).fillna(0).astype(int)

# Encode 'alert'
alert_mapping = {
    'green': 1,
    'yellow': 2,
    'orange': 3,
    'red': 4,
    'none': 0
}
clean_earthquake_copy['alert_encoded'] = clean_earthquake_copy['alert'].fillna('none').map(alert_mapping)

# Encode 'magnitude_category'
magnitude_category_mapping = {
    'minor': 1,
    'light': 2,
    'moderate': 3,
    'strong': 4,
    'major': 5,
    'great': 6
}
clean_earthquake_copy['magnitude_category_encoded'] = (
    clean_earthquake_copy['magnitude_category']
    .fillna('unknown')
    .map(magnitude_category_mapping)
    .fillna(0)
    .astype(int)
)

clean_earthquake_copy.head()
```

```python

# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_auc_score

# Display options
pd.set_option('display.max_columns', 100)

# Load cleaned dataset
DATA_PATH = "/mnt/data/cleaned_asteroid_data.csv"  # ensure this file exists in the environment
df = pd.read_csv(DATA_PATH)

print("Rows:", len(df))
df.head()

# Parse detection_time (HH:MM:SS) into numerical features
def parse_hms_to_seconds(t):
    try:
        h, m, s = str(t).split(":")
        h = int(float(h)) % 24
        m = int(float(m)) % 60
        s = int(float(s)) % 60
        return h*3600 + m*60 + s
    except Exception:
        return np.nan

df["detection_seconds"] = df["detection_time"].apply(parse_hms_to_seconds)

# Drop rows with missing critical fields if any
df = df.dropna(subset=["magnitude", "impact_depth_km", "latitude", "longitude", "detection_seconds"])
print("Rows after basic NA drop:", len(df))

df[["asteroid_id","detection_time","detection_seconds","magnitude","impact_depth_km","latitude","longitude","severity_category","magnitude_type","tsunami_flag"]].head()

# Features & label
numeric_features = ["magnitude", "impact_depth_km", "latitude", "longitude", "detection_seconds"]
categorical_features = ["severity_category", "magnitude_type"]

X = df[numeric_features + categorical_features]
y = df["tsunami_flag"].astype(int)  # proxy label for "impact consequence observed"

# Preprocess: scale numeric, one-hot categorical
numeric_transformer = Pipeline(steps=[
    ("scaler", StandardScaler())
])

categorical_transformer = OneHotEncoder(handle_unknown="ignore")

preprocess = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features),
        ("cat", categorical_transformer, categorical_features)
    ]
)

# Model
clf = LogisticRegression(max_iter=200)

pipe = Pipeline(steps=[("preprocess", preprocess),
                      ("model", clf)])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)

pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)
y_proba = pipe.predict_proba(X_test)[:,1]

print(classification_report(y_test, y_pred))
print("ROC AUC:", roc_auc_score(y_test, y_proba))

# Predict on full dataset
df["impact_probability"] = pipe.predict_proba(X)[:,1]

# Derive a simple "Likely Impact Window" +/- 1 hour around detection_time (placeholder)
def seconds_to_hms(sec):
    sec = int(sec) % (24*3600)
    h = sec // 3600
    m = (sec % 3600) // 60
    s = sec % 60
    return f"{h:02d}:{m:02d}:{s:02d}"

window_seconds = 3600  # 1 hour window (placeholder)
df["impact_window_start"] = (df["detection_seconds"] - window_seconds).apply(seconds_to_hms)
df["impact_window_end"]   = (df["detection_seconds"] + window_seconds).apply(seconds_to_hms)

# Basic info bundle per object
summary_cols = [
    "asteroid_id","magnitude","magnitude_type","severity_category",
    "latitude","longitude","impact_depth_km","detection_time",
    "impact_window_start","impact_window_end","impact_probability"
]

predictions = df[summary_cols].copy()
predictions = predictions.sort_values("impact_probability", ascending=False).reset_index(drop=True)

predictions.head(10)

# Predict on full dataset
df["impact_probability"] = pipe.predict_proba(X)[:,1]

# Derive a simple "Likely Impact Window" +/- 1 hour around detection_time (placeholder)
def seconds_to_hms(sec):
    sec = int(sec) % (24*3600)
    h = sec // 3600
    m = (sec % 3600) // 60
    s = sec % 60
    return f"{h:02d}:{m:02d}:{s:02d}"

window_seconds = 3600  # 1 hour window (placeholder)
df["impact_window_start"] = (df["detection_seconds"] - window_seconds).apply(seconds_to_hms)
df["impact_window_end"]   = (df["detection_seconds"] + window_seconds).apply(seconds_to_hms)

# Basic info bundle per object
summary_cols = [
    "asteroid_id","magnitude","magnitude_type","severity_category",
    "latitude","longitude","impact_depth_km","detection_time",
    "impact_window_start","impact_window_end","impact_probability"
]

predictions = df[summary_cols].copy()
predictions = predictions.sort_values("impact_probability", ascending=False).reset_index(drop=True)

predictions.head(10)

# Visual: probability distribution
plt.figure(figsize=(6,4))
plt.hist(predictions["impact_probability"], bins=30)
plt.title("Predicted Impact Probability (Proxy Model)")
plt.xlabel("Probability")
plt.ylabel("Count")
plt.show()
```
